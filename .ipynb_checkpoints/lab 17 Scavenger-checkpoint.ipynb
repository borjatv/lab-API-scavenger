# Challenge 3) Hidden cold joke

```python
params = {'accept': 'application/vnd.github.v3+json',
          'path': '15024'  
}
```

```python
cold = requests.get(f'{url}repos/ironhack-datalabs/scavenger/contents/', headers=headers)
```

```python
cold.json()
```

```python
dirs = pd.json_normalize(cold.json())
dirs
```

```python
list_path = list(dirs)['path'])
```

```python
allfiles = pd.DataFrame()
```

```python
for x in list_path:
    coldnew = requests.get(f'{url}repos/ironhack-datalabs/scavenger/contents/{x}', headers=headers)
    allfiles = allfiles.append(pd.json_normalize(coldnew.json()))
    
```

```python
allfiles
```

```python
requests.get(f'{url}repos/ironhack-datalabs/scavenger/contents/15024/.0006.scavengerhunt', headers=headers).json()
```

```python
pd.json_normalize(requests.get(f'{url}repos/ironhack-datalabs/scavenger/contents/15024/.0006.scavengerhunt', headers=headers).json())
```

```python
fileinfo = pd.DataFrame()
```

```python
for i,c in allfiles.iterrows():
    if c['name'].endswith('scavengerhunt'):
        x = pd.json_normalize(requests.get(f'{url}repos/ironhack-datalabs/scavenger/contents/{c["path"]}', headers=headers).json())
        fileinfo = fileinfo.append(x)
        
```

```python
fileinfo = fileinfo.sort_values('name')
```

```python
fileinfo[['name', 'content']]
```

```python
jokelist = fileinfo['content'].to_list()
```

```python
newjokelist = [a.replace('\n','') for a in jokelist]
```

```python
newjokelist
```

```python
mystery = ''
for a in range(len(newjokelist)):
    mystery = mystery + jokelist[a]
        

    
print(mystery)
```

```python
import base64
```

```python
for a in newjokelist:
    print(str(base64.b64decode(a), 'utf-8'))
```
